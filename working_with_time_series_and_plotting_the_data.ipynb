#Rolling Window Analysis

Application of any method or model can be used repeatedly to the sub-data sets or sub-series in your full data set or series.

It can be applied to time series data that fluctuates rapidly.

Some use cases:

1. Stock Market Analysis
2. sales/Revenue forecast
3. Power Load prediction
4. Crop Yield Prediction



```python
temp.rolling(window=3).mean() #ekhane ager duita k niye avg kore
```

```python
temp.rolling(window=3,  center= True).mean() #ager o porer value niye avg
```

```python
temp.plot()
```

```python
temp.plot(kind='line',title='hourly_temp', style='.-',color='r',mfc='y', ms= 8) #can not demonstrat the histogram
```
