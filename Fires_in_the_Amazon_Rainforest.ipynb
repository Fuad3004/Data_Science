#All in One Till Now

```python
amazon_fire="https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/amazon_fires.csv"

df= pd.read_csv(amazon_fire, encoding="ISO-8859-1")

new_column= { 'ano' : 'year',
             'estado': 'state',
              'mes' : 'mouth',
              'numero' : 'number_of_fires',
              'encontro': 'date'}

df.rename(columns=new_column, inplace= True)
df['number_of_fires'] = df['number_of_fires'].str.strip(" Fires")
df['number_of_fires'] = df["number_of_fires"].str.replace('','0').astype(float)

df_copy=df.copy()
df.head()
```

#Handling missing data

```python
df.isnull().sum()
```

# Drop rows with NaN values

```python
df= df.dropna()

df= df.reset_index() # reset's row indexes in case any rows were dropped

df.head()
```

```python
df.isnull().sum()
```

so now there are no null values.

```python
df['number_of_fires'].isnull()
```

What do to with missing data?
Remove them via .dropna(axis=0) or,
Replace them with some arbitary number (e.g. an average) , or
Replace them zeros, or Forward Fill (ffill) or Back Fill (backfill)

```python
# Using fillna with zeros

df_copy['number_of_fires'].fillna(0).head()
```
